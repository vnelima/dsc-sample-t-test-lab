```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def one_sample_ttest(sample, popmean, alpha):
    # Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    n = len(sample)
    
    # Calculate the t-statistic
    t_statistic = (sample_mean - popmean) / (sample_std / np.sqrt(n))
    
    # Calculate the degrees of freedom
    df = n - 1
    
    # Calculate the critical t-value and p-value
    critical_t_value = stats.t.ppf(1 - alpha / 2, df)
    p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
    
    # Print the results
    print(f'T-statistic: {t_statistic:.4f}')
    print(f'Critical t-value: {critical_t_value:.4f}')
    print(f'P-value: {p_value:.4f}')
    
    # Determine if results are significant
    if p_value < alpha:
        print('Result is significant at alpha =', alpha)
    else:
        print('Result is not significant at alpha =', alpha)
    
    # Create a visualization to check for normality
    plt.figure(figsize=(10, 6))
    sns.histplot(sample, bins=20, kde=True)
    plt.axvline(popmean, color='r', linestyle='dashed', linewidth=2, label='Population Mean')
    plt.axvline(sample_mean, color='b', linestyle='dashed', linewidth=2, label='Sample Mean')
    plt.title('Sample Distribution with Population and Sample Mean')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Example usage
sample_data = np.random.normal(loc=50, scale=10, size=30)
population_mean = 52
alpha = 0.05

one_sample_ttest(sample_data, population_mean, alpha)
```
